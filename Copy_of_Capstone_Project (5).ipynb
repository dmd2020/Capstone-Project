```
import pandas as pd
missing_values = 'nan'
df = pd.read_csv('https://query.data.world/s/s5jfyzfsgvjrc4nwg6ooe3ks5p3h4p', na_values = missing_values)
df= df[(df['Question']=='Cancer of the colon and rectum (colorectal), incidence') &  (df['DataValueType']=='Average Annual Age-adjusted Rate')]
df['DataValue'] = df['DataValue'].astype(float)
df = df.groupby(['LocationAbbr'])['DataValue'].mean().sort_values(ascending= False)
df


```

2008-2012 Average Annual Age of Colorectal Cancer Incidences By State

```
import pandas as pd
missing_values = 'nan'
df = pd.read_csv('https://query.data.world/s/s5jfyzfsgvjrc4nwg6ooe3ks5p3h4p', na_values = missing_values)
df= df[(df['Question']=='Cancer of the colon and rectum (colorectal), incidence') &  (df['DataValueType']=='Average Annual Number')]
df['DataValue'] = df['DataValue'].astype(float)
df = df.groupby(['LocationAbbr'])['DataValue'].mean().sort_values(ascending= False)
df
```
